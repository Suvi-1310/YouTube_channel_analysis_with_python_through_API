```python
df = pd.DataFrame(columns=["video_id","video_title","upload_date","view_count"])
```

###creating_second_loop

```python
for video in response['items']:
  if video['id']['kind'] == "youtube#video":                                       

     video_id = video['id']['videoId']
     video_title = video['snippet']['title']
     upload_date = video['snippet']['publishedAt']
     upload_date = str(upload_date).split("T")[0]

      #Second API call for view, like, dislike, comment counts
     url_video_status = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+API_KEY
     response_video_status = requests.get(url_video_status).json()

     view_count = response_video_status['items'][0]['statistics']['viewCount']
    

     df = df.append({'video_id' : video_id, 'video_title' :video_title, 'upload_date' :upload_date, 'view_count' :view_count}, ignore_index = True)
```
